```python
import pandas as pd
import requests
import urllib.parse

basedir = "/home/romi/Documents/Research/OntoEngineerings/Clinical_Trials_Outcomes/git/autocoo/data/"
url = 'https://clinicaltrials.gov/api/v2/studies'

diabetes_data = pd.read_csv(basedir+"data/diatebes-studies.csv",usecols=[0,1,2,3],header=0)
cancer_data = pd.read_csv(basedir+"data/cancer-studies.csv",usecols=[0,1,2,3,4],header=0)
# print(diabetes_data)
```

## Get Study information without ID (Diabetes)

```python
diabetes_data_long = []

for ind,row in diabetes_data.iterrows():
    long_row = row
    title = row[0]
    query_term = row[0]
    print(query_term)
    
    if '"' in query_term:
        query_term = query_term.split('"')[0]
    
    params = {'query.cond': 'diabetes', 'query.term': '"'+query_term+'"'}
    response = requests.get(url+"?" + urllib.parse.urlencode(params))
    jresponse = response.json()
    
    for study in jresponse['studies']:
        if (study['protocolSection']['identificationModule']['briefTitle'] == title):
    
            # protocolSection -> identificationModule -> nctId
            nct_id = study['protocolSection']['identificationModule']['nctId']
            long_row['nct_id'] = nct_id

            # protocolSection -> statusModule -> startDateStruct -> date
            if ('startDateStruct' in study['protocolSection']['statusModule'].keys()):
                date = study['protocolSection']['statusModule']['startDateStruct']['date']
                long_row['startDate'] = date
            else:
                long_row['startDate'] = 'na'

            # protocolSection -> statusModule -> overallStatus
            status = study['protocolSection']['statusModule']['overallStatus']
            long_row['status'] = status

            #  protocolSection -> designModule -> studyType
            stype = study['protocolSection']['designModule']['studyType']
            long_row['study_type'] = stype

            # protocolSection -> designModule -> phases
            if ('phases' in study['protocolSection']['designModule'].keys()):
                phases = study['protocolSection']['designModule']['phases']
                long_row['phases'] = phases
#                 if len(phases)>1:
#                     print(phases)
#                     for phase in phases:
#                         long_row['phases']

            else:
                long_row['phases'] = ['NA']

            
            diabetes_data_long.append(long_row)
#             print(len(diabetes_data_long))


# print(dd)
```

```python
dd = pd.DataFrame.from_dict(diabetes_data_long)
# dd.loc[dd["phases"] == "na", "phases"] = ['NA']
# dd.loc[dd["phases"] == "['NA']", "phases"] = ['NA']
dd.to_csv(basedir+"diabetes-studies-addinfo.csv",index=False)
# dd = pd.read_csv(basedir+"data/diabetes-studies-addinfo.csv")
```

```python
dd_manual = dd.head(60)
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True,figsize=(10, 4))
fig.suptitle('Diabetes: 60 CTs used for the manual engineering')
# ax1.plot(dd_manual['phases'].value_counts())
# ax2.plot(dd_manual['phases'].value_counts())

dd_manual['phases'].value_counts().plot(kind='bar', ax=ax1)
dd_manual['status'].value_counts().plot(kind='bar', ax=ax2)
dd_manual['study_type'].value_counts().plot(kind='bar', ax=ax3)

# fig.show()
plt.savefig(basedir+"diabetes-ctinfo-manual.pdf",bbox_inches="tight")
```

## Get Study information by ID (Gastrointestinal Cancer)

```python
cancer_data_long = []

for ind,row in cancer_data.iterrows():
    long_row = row
    title = row[0]
    nct_id = row[0]
    print(nct_id)
    
    response = requests.get(url +"/"+nct_id)
    study = response.json()
    
    if ('startDateStruct' in study['protocolSection']['statusModule'].keys()):
        date = study['protocolSection']['statusModule']['startDateStruct']['date']
        long_row['startDate'] = date
    else:
        long_row['startDate'] = 'na'

    # protocolSection -> statusModule -> overallStatus
    status = study['protocolSection']['statusModule']['overallStatus']
    long_row['status'] = status

    #  protocolSection -> designModule -> studyType
    stype = study['protocolSection']['designModule']['studyType']
    long_row['study_type'] = stype

    # protocolSection -> designModule -> phases
    if ('phases' in study['protocolSection']['designModule'].keys()):
        phases = study['protocolSection']['designModule']['phases']
        long_row['phases'] = phases

    else:
        long_row['phases'] = ['NA']

    cancer_data_long.append(long_row)
```

```python
cd = pd.DataFrame.from_dict(cancer_data_long)
cd.to_csv(basedir+"cancer-studies-addinfo.csv",index=False)
```

```python
cd_manual = cd.sample(n=90)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True,figsize=(10, 4))
fig.suptitle('GI Cancer: 90 random CTs used for the manual engineering')

cd_manual['phases'].value_counts().plot(kind='bar', ax=ax1)
cd_manual['status'].value_counts().plot(kind='bar', ax=ax2)
cd_manual['study_type'].value_counts().plot(kind='bar', ax=ax3)

# fig.show()
plt.savefig(basedir+"cancer-ctinfo-manual.pdf",bbox_inches="tight")
```

# Test set for document tagging

retrieving a list of clinical trials that was not used in the process of semi-/automatic or manual engineering of the ontologies

```python
import pandas as pd
import requests
import urllib.parse


basedir = "/home/romi/Documents/Research/OntoEngineerings/Clinical_Trials_Outcomes/git/autocoo/data/"
url = 'https://clinicaltrials.gov/api/v2/studies'


def retrieve_ct(term):
    params = {'format': 'csv', 'query.cond': term, 'fields': 'NCT Number', 'pageSize': 1000}
    response = requests.get(url+"?" + urllib.parse.urlencode(params))

    # print(response)
    nct_set = set(response.text.split()[2:])
    nextpagetoken = response.headers['x-next-page-token']
    # print(nextpagetoken)

    while 'x-next-page-token' in response.headers.keys():
        nextpagetoken = response.headers['x-next-page-token']
        response = requests.get(url+"?" + urllib.parse.urlencode(params))
    #     print(response)
        params = {'format': 'csv', 'query.cond': term, 'fields': 'NCT Number', 'pageSize': 1000, 'pageToken': nextpagetoken} 
        res = response.text.split()[2:]
#         print(nextpagetoken,len(res))
        nct_set |= set(res)
    
    return nct_set
```

### Diabetes

```python
dd = pd.read_csv(basedir+"diabetes-studies-addinfo.csv")
dd.keys()
```

```python
# https://clinicaltrials.gov/api/v2/studies?format=csv&query.cond=diabetes&fields=NCT+Number&pageSize=1000
term = 'diabetes'

nct_dd = retrieve_ct(term)

print("retrieved clinical trials:",len(nct_dd))
used_dd = set(dd['nct_id'])
print("used clinical trials in modelling:", len(used_dd))
test_dd = nct_dd - used_dd
print("can be used for testing:", len(test_dd))
print("train/test split:", len(used_dd)/len(nct_dd), "/", len(test_dd)/len(nct_dd))
```

```python
# write the complete list of clinical trial ids to file
with open(basedir+"diabetes-ct-ids-all.txt", 'w') as f:
    for item in nct_dd:
        f.write("%s\n" % item)
print("done")

# write the train set list of clinical trial ids to file
with open(basedir+"diabetes-ct-ids-train.txt", 'w') as f:
    for item in used_dd:
        f.write("%s\n" % item)
print("done")

# write the test set list of clinical trial ids to file
with open(basedir+"diabetes-ct-ids-test.txt", 'w') as f:
    for item in test_dd:
        f.write("%s\n" % item)
print("done")
```
