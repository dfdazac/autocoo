```python
import pandas as pd
import requests
import urllib.parse

basedir = "/home/romi/Documents/Research/OntoEngineerings/Clinical_Trials_Outcomes/git/autocoo/data/"
url = 'https://clinicaltrials.gov/api/v2/studies'

diabetes_data = pd.read_csv(basedir+"data/diatebes-studies.csv",usecols=[0,1,2,3],header=0)
cancer_data = pd.read_csv(basedir+"data/cancer-studies.csv",usecols=[0,1,2,3,4],header=0)
# print(diabetes_data)
```

## Get Study information without ID (Diabetes)

```python
diabetes_data_long = []

for ind,row in diabetes_data.iterrows():
    long_row = row
    title = row[0]
    query_term = row[0]
    print(query_term)
    
    if '"' in query_term:
        query_term = query_term.split('"')[0]
    
    params = {'query.cond': 'diabetes', 'query.term': '"'+query_term+'"'}
    response = requests.get(url+"?" + urllib.parse.urlencode(params))
    jresponse = response.json()
    
    for study in jresponse['studies']:
        if (study['protocolSection']['identificationModule']['briefTitle'] == title):
    
            # protocolSection -> identificationModule -> nctId
            nct_id = study['protocolSection']['identificationModule']['nctId']
            long_row['nct_id'] = nct_id

            # protocolSection -> statusModule -> startDateStruct -> date
            if ('startDateStruct' in study['protocolSection']['statusModule'].keys()):
                date = study['protocolSection']['statusModule']['startDateStruct']['date']
                long_row['startDate'] = date
            else:
                long_row['startDate'] = 'na'

            # protocolSection -> statusModule -> overallStatus
            status = study['protocolSection']['statusModule']['overallStatus']
            long_row['status'] = status

            #  protocolSection -> designModule -> studyType
            stype = study['protocolSection']['designModule']['studyType']
            long_row['study_type'] = stype

            # protocolSection -> designModule -> phases
            if ('phases' in study['protocolSection']['designModule'].keys()):
                phases = study['protocolSection']['designModule']['phases']
                long_row['phases'] = phases
#                 if len(phases)>1:
#                     print(phases)
#                     for phase in phases:
#                         long_row['phases']

            else:
                long_row['phases'] = ['NA']

            
            diabetes_data_long.append(long_row)
#             print(len(diabetes_data_long))


# print(dd)
```

```python
dd = pd.DataFrame.from_dict(diabetes_data_long)
# dd.loc[dd["phases"] == "na", "phases"] = ['NA']
# dd.loc[dd["phases"] == "['NA']", "phases"] = ['NA']
dd.to_csv(basedir+"diabetes-studies-addinfo.csv",index=False)
# dd = pd.read_csv(basedir+"data/diabetes-studies-addinfo.csv")
```

# Test set for document tagging

retrieving a list of clinical trials that was not used in the process of semi-/automatic or manual engineering of the ontologies

```python
import pandas as pd
import requests
import urllib.parse


basedir = "/home/romi/Documents/Research/OntoEngineerings/Clinical_Trials_Outcomes/git/autocoo/data/"
url = 'https://clinicaltrials.gov/api/v2/studies'


def retrieve_ct(term):
    params = {'format': 'csv', 'query.cond': term, 'fields': 'NCT Number', 'pageSize': 1000}
    response = requests.get(url+"?" + urllib.parse.urlencode(params))

    # print(response)
    nct_set = set(response.text.split()[2:])
    nextpagetoken = response.headers['x-next-page-token']
    # print(nextpagetoken)

    while 'x-next-page-token' in response.headers.keys():
        nextpagetoken = response.headers['x-next-page-token']
        response = requests.get(url+"?" + urllib.parse.urlencode(params))
    #     print(response)
        params = {'format': 'csv', 'query.cond': term, 'fields': 'NCT Number', 'pageSize': 1000, 'pageToken': nextpagetoken} 
        res = response.text.split()[2:]
#         print(nextpagetoken,len(res))
        nct_set |= set(res)
    
    return nct_set
```

# Train/Test and 5 fold CV splits

```python
from sklearn.model_selection import KFold
import os

with open("cancer-ct-ids-all.txt", 'r') as f:
    data = f.read()
    cd = data.split()

with open("diabetes-ct-ids-all.txt", 'r') as f:
    data = f.read()
    dd = data.split()

print(dd)
```

```python
from requests.exceptions import ConnectionError



def get_extended_training_data(data):
    data_long = pd.DataFrame()
    print("querying for clinical trial info")
    for nct_id in data:
        # print(nct_id)
        long_row = {}
        long_row['nct_id'] = nct_id
        try:
            response = requests.get(url +"/"+nct_id)
               
            study = response.json()
            
            # print(study['protocolSection']['outcomesModule'].keys())
            
            # break
            if ('outcomesModule' in study['protocolSection'].keys()):
                for key in study['protocolSection']['outcomesModule'].keys():
                    for measure in study['protocolSection']['outcomesModule'][key]:
                        if ('measure') in measure.keys():
                            long_row[key] = measure['measure']

            
            if ('startDateStruct' in study['protocolSection']['statusModule'].keys()):
                date = study['protocolSection']['statusModule']['startDateStruct']['date']
                long_row['startDate'] = date
            else:
                long_row['startDate'] = 'na'

            # protocolSection -> statusModule -> overallStatus
            status = study['protocolSection']['statusModule']['overallStatus']
            long_row['status'] = status

            #  protocolSection -> designModule -> studyType
            stype = study['protocolSection']['designModule']['studyType']
            long_row['study_type'] = stype

            # protocolSection -> designModule -> phases
            if ('phases' in study['protocolSection']['designModule'].keys()):
                phases = study['protocolSection']['designModule']['phases']
                long_row['phases'] = phases

            else:
                long_row['phases'] = ['NA']
            
            data_long = data_long.append(long_row,ignore_index=True)
            
        except ConnectionError as e:    # This is the correct syntax
            r = "No response for " + nct_id
            print(r)

    print("done querying for clinical trial info")
    return data_long
    
```

```python


def split_data(term,data):
    kf = KFold(n_splits=5,shuffle=True)
    i = 0
    
    for train, test in kf.split(data):
        print("%s %s" % (len(train),len(test)))
        folder = os.path.join(term,"cross-validation","split_"+str(i))
        os.makedirs(folder,exist_ok=True)

        train_data = [data[i] for i in train]
        test_data = [data[i] for i in test]
        train_data_long = get_extended_training_data(train_data)
        
        train_data_long.to_csv(os.path.join(folder,"train_addinfo.csv"))
        
        with open(os.path.join(folder,"train.txt"), 'w') as f:
            for item in train_data:
                f.write("%s\n" % item)

        with open(os.path.join(folder,"test.txt"), 'w') as f:
            for item in test_data:
                f.write("%s\n" % item)

        

        i = i+1
```
